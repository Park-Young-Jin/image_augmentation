```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
result_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/졸업논문/졸업논문실험결과3.csv')
```

```python
#result_data_for_plot_all = result_data
result_data_for_plot_cifar10 = result_data[result_data['image_set']=='cifar10'] #.loc[:,['hist','cos','conv','acc_3']]
result_data_for_plot_cifar10_subset_25000 = result_data[result_data['image_set']=='cifar10_subset_25000'] #.loc[:,['hist','cos','conv','acc_3']]
result_data_for_plot_cifar10_subset_12850 = result_data[result_data['image_set']=='cifar10_subset_12850'] #.loc[:,['hist','cos','conv','acc_3']]
result_data_for_plot_cifar10_subset_5000 = result_data[result_data['image_set']=='cifar10_subset_5000'] #.loc[:,['hist','cos','conv','acc_3']]
#result_data_for_plot_size32 = result_data[result_data['size']==32] #.loc[:,['hist','cos','conv','acc_3']]


result_data_for_plot_imagenet32 = result_data[result_data['image_set']=='imagenet32'] #.loc[:,['hist','cos','conv','acc_3']]
result_data_for_plot_imagenet64 = result_data[result_data['image_set']=='imagenet64'] #.loc[:,['hist','cos','conv','acc_3']]
```

```python
# 그래프에 잘 보이도록 일부 데이터만 추출하기

#result_data_for_plot_cifar10 = result_data_for_plot_cifar10[result_data_for_plot_cifar10['1']==1]
#result_data_for_plot_cifar10_subset_25000 = result_data_for_plot_cifar10_subset_25000[result_data_for_plot_cifar10_subset_25000['1']==1] 
#result_data_for_plot_cifar10_subset_12850 = result_data_for_plot_cifar10_subset_12850[result_data_for_plot_cifar10_subset_12850['1']==1] 
#result_data_for_plot_cifar10_subset_5000 = result_data_for_plot_cifar10_subset_5000[result_data_for_plot_cifar10_subset_5000['1']==1]


#result_data_for_plot_cifar10 = result_data_for_plot_cifar10[result_data_for_plot_cifar10['2']!=15]
#result_data_for_plot_cifar10_subset_25000 = result_data_for_plot_cifar10_subset_25000[result_data_for_plot_cifar10_subset_25000['2']!=15] 
#result_data_for_plot_cifar10_subset_12850 = result_data_for_plot_cifar10_subset_12850[result_data_for_plot_cifar10_subset_12850['2']!=15] 
#result_data_for_plot_cifar10_subset_5000 = result_data_for_plot_cifar10_subset_5000[result_data_for_plot_cifar10_subset_5000['2']!=15]
```

```python
cifar10_base = result_data_for_plot_cifar10[(result_data_for_plot_cifar10['1']==0) & (result_data_for_plot_cifar10['2']==0) & (result_data_for_plot_cifar10['3']==0)]['acc_3']
cifar10_subset_base_25000 = result_data_for_plot_cifar10_subset_25000[(result_data_for_plot_cifar10_subset_25000['1']==0) & (result_data_for_plot_cifar10_subset_25000['2']==0) & (result_data_for_plot_cifar10_subset_25000['3']==0)]['acc_3']
cifar10_subset_base_12850 = result_data_for_plot_cifar10_subset_12850[(result_data_for_plot_cifar10_subset_12850['1']==0) & (result_data_for_plot_cifar10_subset_12850['2']==0) & (result_data_for_plot_cifar10_subset_12850['3']==0)]['acc_3']
cifar10_subset_base_5000 = result_data_for_plot_cifar10_subset_5000[(result_data_for_plot_cifar10_subset_5000['1']==0) & (result_data_for_plot_cifar10_subset_5000['2']==0) & (result_data_for_plot_cifar10_subset_5000['3']==0)]['acc_3']

imagenet32_base = result_data_for_plot_imagenet32[(result_data_for_plot_imagenet32['1']==0) & (result_data_for_plot_imagenet32['2']==0) & (result_data_for_plot_imagenet32['3']==0)]['acc_3']
imagenet64_base = result_data_for_plot_imagenet64[(result_data_for_plot_imagenet64['1']==0) & (result_data_for_plot_imagenet64['2']==0) & (result_data_for_plot_imagenet64['3']==0)]['acc_3']
```

```python
#성능의 절대값은 다를 수 있음. 상대적 비교를 위해 정규화 처리
#같은 데이터 두 번 처리한 것을 1로 두고 상대값으로 처리


result_data_for_plot_cifar10['acc_3'] = result_data_for_plot_cifar10['acc_3']/float(cifar10_base)
result_data_for_plot_cifar10_subset_25000['acc_3'] = result_data_for_plot_cifar10_subset_25000['acc_3']/float(cifar10_subset_base_25000)
result_data_for_plot_cifar10_subset_12850['acc_3'] = result_data_for_plot_cifar10_subset_12850['acc_3']/float(cifar10_subset_base_12850)
result_data_for_plot_cifar10_subset_5000['acc_3'] = result_data_for_plot_cifar10_subset_5000['acc_3']/float(cifar10_subset_base_5000)


result_data_for_plot_imagenet32['acc_3'] = result_data_for_plot_imagenet32['acc_3']/float(imagenet32_base)
result_data_for_plot_imagenet64['acc_3'] = result_data_for_plot_imagenet64['acc_3']/float(imagenet64_base)
```

```python
result_data_for_plot_cifar10.sort_values('acc_3')
result_data_for_plot_cifar10_subset_25000.sort_values('acc_3')
result_data_for_plot_cifar10_subset_12850.sort_values('acc_3')
result_data_for_plot_cifar10_subset_5000.sort_values('acc_3')

result_data_for_plot_imagenet32.sort_values('acc_3')
result_data_for_plot_imagenet64.sort_values('acc_3')
```

```python
def graph_each(dataset_name,sim_type,name,x_min):
    plt.plot(sim_type,  # x
            'acc_3',  # y
            data=dataset_name, 
            linestyle='none',
            marker='o', 
            markersize=10,
            color='blue', 
            alpha=0.5)

    plt.xlabel('similarity(DataSet : '+name+')', fontsize=14)
    plt.ylabel('accuracy', fontsize=14)
    plt.grid()
    plt.xlim(x_min,1.05) #0.7
    plt.ylim(0.9,1.05)
    
    #plt.xlim(0,1.05)
    #plt.ylim(0.9,1.05)
    plt.show()
```

```python
graph_each(result_data_for_plot_cifar10,'conv','CIFAR10_all',0.7)
graph_each(result_data_for_plot_cifar10_subset_25000,'conv','CIFAR10_25000',0.7)
graph_each(result_data_for_plot_cifar10_subset_12850,'conv','CIFAR10_12850',0.7)
graph_each(result_data_for_plot_cifar10_subset_5000,'conv','CIFAR10_5000',0.7)

graph_each(result_data_for_plot_imagenet32,'conv','ImageNet32',0.7)
graph_each(result_data_for_plot_imagenet64,'conv','ImageNet64',0.7)
```
