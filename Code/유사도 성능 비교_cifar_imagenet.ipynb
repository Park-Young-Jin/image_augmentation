```python
def graph_each(dataset_name):
    plt.plot('hist',  # x
            'acc_3',  # y
            data=dataset_name, 
            linestyle='none',
            marker='o', 
            markersize=10,
            color='blue', 
            alpha=0.5)

    plt.xlabel('similarity(histogram)', fontsize=14)
    plt.ylabel('accuracy', fontsize=14)
    plt.show()

    plt.plot('cos',  # x
            'acc_3',  # y
            data=dataset_name, 
            linestyle='none',
            marker='o', 
            markersize=10,
            color='red', 
            alpha=0.5)

    plt.xlabel('similarity(cosine)', fontsize=14)
    plt.ylabel('accuracy', fontsize=14)
    plt.show()

    plt.plot('conv',  # x
            'acc_3',  # y
            data=dataset_name, 
            linestyle='none',
            marker='o', 
            markersize=10,
            color='green', 
            alpha=0.5)

    #plt.title('성능과 유사도 관계', fontsize=20)
    plt.xlabel('similarity(CNN)', fontsize=14)
    plt.ylabel('accuracy', fontsize=14)
    plt.show()

    plt.plot('conv2',  # x
            'acc_3',  # y
            data=dataset_name, 
            linestyle='none',
            marker='o', 
            markersize=10,
            color='yellow', 
            alpha=0.5)

    #plt.title('성능과 유사도 관계', fontsize=20)
    plt.xlabel('similarity(CNN+PCA)', fontsize=14)
    plt.ylabel('accuracy', fontsize=14)
    plt.show()    
```

```python
def model_train(x,y,train_dataset,test_dataset,model_algo): 
    y_col = y #["acc_3"]
    x_cols = x #["hist","cos","conv"]

    X = train_dataset[x_cols].values #result_data_for_plot_cifar10[x_cols].values
    y = train_dataset[y_col].values#result_data_for_plot_cifar10[y_col].values
    
    if model_algo == "LinearRegreesion":
        model = linear_model.LinearRegression().fit(X, y)
    #elif model_algo == "Ridge":
    #    model = linear_model.Ridge(alpha=0.0001).fit(X, y)    
    #elif model_algo == "Lasso":
    #    model = linear_model.Lasso(alpha=0.0001).fit(X, y)  
    #elif model_algo == "BayesianRidge":
    #    model = linear_model.BayesianRidge().fit(X, y)    
    elif model_algo == "PolyRegreesion":           
        #poly_features = PolynomialFeatures(degree=2,include_bias=False)
        #x_poly = poly_features.fit_transform(X)
        #model = linear_model.LinearRegression().fit(x_poly,y)
        model = Pipeline([
                           ("poly_features",PolynomialFeatures(degree=2,include_bias=True)),
                           ("lin_reg",linear_model.LinearRegression())
                         ])
        model.fit(X,y)
    elif model_algo == "DecisionTree":   
        model = tree.DecisionTreeRegressor().fit(X, y) 
    elif model_algo == "SVR":    
        #model = svm.SVR(C=0.00001,gamma = 0.0001,epsilon = 0.00001).fit(X, y)
        model = Pipeline([
                           ('std_scaler',StandardScaler()),
                           ("SVM_reg",svm.SVR(kernel = 'linear',C=0.00001,gamma = 0.0001,epsilon = 0.00001))
                         ])
        model.fit(X,y)
        

    print(model_algo)
    print(x_cols)
        #모델별로 추가로 볼 정보들
    if model_algo == "LinearRegreesion":
        print(model.coef_)

       

    for i in range(len(test_dataset)):
        data_lenth = len(test_dataset[i][x_cols]) 
        print('rmse : ',(sum((model.predict(test_dataset[i][x_cols].values).reshape(data_lenth,1)-test_dataset[i][y_col].values)**2)/data_lenth)**(1/2) )    
        #print(model.score(test_dataset[i][x_cols].values, test_dataset[i][y_col].values))    

        plot_x = np.array(range(len(test_dataset[i])))
        plot_y = model.predict(test_dataset[i][x_cols].values)
        plt.plot(plot_x, plot_y, color = "red", marker = "o", label = "predict")
        plt.legend()

        plot_x = np.array(range(len(test_dataset[i])))
        plot_y = test_dataset[i][y_col].values
        plt.plot(plot_x, plot_y, color = "blue", marker = "o", label = "real")
        
        plt.xlabel("index")
        plt.ylabel("acc")
        
        plt.legend()
        plt.show()





```
